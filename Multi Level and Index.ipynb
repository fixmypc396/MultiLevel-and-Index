# Multi-Level and Index examples

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/steven/Jupyter/archive/athlete_events.csv')

df.head()
```

```python
# I want all games from 2012
df.loc[df['Year'] == 2012]
```

```python
# Set the Year as the index
df = df.set_index('Year')
```

```python
# Since we reset the index I can pull data using loc
df.loc[1992]
```

```python
# Lets set the index by Year and season
# We have to rest the index 
df = df.reset_index()

# Verify
df.head()
```

```python
df = df.set_index(['Year', 'Season'])

df.head() # See how the year and season column are on the left and are a multiIndex
```

```python
# Sort the index first!! It is sort by an integer and then by string
df = df.sort_index()
# Now we can search by year and Season
df = df.loc[('1992','Summer')]

df.head()
```

```python
# Lets do another index example with Year, Season, and Sport
df = df.reset_index()
df = df.set_index(['Year', 'Season', 'Sport'])
df = df.sort_index()
df.head()
```

```python
# Lets retreive by year
df.loc[1992] # Remember to use [] after .loc
```

```python
# Do one by the year and season and sport
df.loc[(1992, 'Summer', 'Archery')]
```

```python
df.xs('Archery', level = 'Sport') # .xs is a cross section and Level is the column index and Archery is a variable.
```

```python
# Let do the average height and weight of archers
df.xs('Archery', level = 'Sport')[['Height', 'Weight']].mean()
```

```python
# Here is how to specify more than one thing with .xs
df.xs((1992, 'Archery'), level = ('Year', 'Sport')) # See how the level for 1992 is Year.
```

```python
df.xs((1992, 'Archery'), level = ('Year', 'Sport'))[['Height', 'Weight']].mean()
```
